```python
from artm_utils.optimizations import *
from artm_utils.calculations import *
from artm_utils.dataset_preparations import *
from artm_utils.regularizers import *
from sklearn.datasets import fetch_20newsgroups
import numpy as np
```

```python
train_n_dw_matrix, token_2_num, num_2_token = prepare_twitter_sentiment140_dataset('/home/tylorn/training.1600000.processed.noemoticon.csv', calc_cooccurences=False)
```

```python
D, W = train_n_dw_matrix.shape
```

```python
T = 20
top_size = 20
```

```python
random_gen = np.random.RandomState(42)
phi_matrix = get_prob_matrix_by_counters(random_gen.uniform(size=(T, W)).astype(np.float64))
theta_matrix = get_prob_matrix_by_counters(np.ones(shape=(D, T)).astype(np.float64))
```

```python
regularization_list = np.zeros(100, dtype=object)
regularization_list[:] = create_reg_lda(0., 0.)
```

```python
phi_base, _ = em_optimization(
    n_dw_matrix=train_n_dw_matrix,
    phi_matrix=phi_matrix,
    theta_matrix=theta_matrix,
    regularization_list=regularization_list,
    iters_count=100,
    params={}
)

for t in xrange(T):
    top = heapq.nlargest(top_size, xrange(W), lambda w: phi_base[t, w])
    print t
    for w in top:
        print '\t', num_2_token[w], phi_base[t, w]
    print ''
```

```python
phi_artm, _ = artm_thetaless_em_optimization(
    n_dw_matrix=train_n_dw_matrix,
    phi_matrix=phi_matrix,
    theta_matrix=theta_matrix,
    regularization_list=regularization_list,
    iters_count=100,
    params={'use_B_cheat': False}
)

for t in xrange(T):
    top = heapq.nlargest(top_size, xrange(W), lambda w: phi_artm[t, w])
    print t
    for w in top:
        print '\t', num_2_token[w], phi_artm[t, w]
    print ''
```

```python
for t_base in xrange(T):
    closest_t_artm = -1
    closest_val = 1e9
    for t_artm in xrange(T):
        dist = np.sum((np.sqrt(phi_base[t_base, :]) - np.sqrt(phi_artm[t_artm, :])) ** 2)
        if dist < closest_val:
            closest_t_artm = t_artm
            closest_val = dist
    print t_base, closest_t_artm, closest_val
    top_base = heapq.nlargest(top_size, xrange(W), lambda w: phi_base[t_base, w])
    top_artm = heapq.nlargest(top_size, xrange(W), lambda w: phi_artm[closest_t_artm, w])
    
    for w1, w2 in zip(top_artm, top_base):
        print '\t{:<15}\t{:<15}'.format(num_2_token[w1], num_2_token[w2])

    print ''
            
```
